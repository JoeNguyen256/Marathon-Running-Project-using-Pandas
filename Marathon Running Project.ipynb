```python
df2['Athlete year of birth'] = df2['Athlete year of birth'].astype('string')
df2['Athlete year of birth'] = df2['Athlete year of birth'].str.split('.').str.get(0)
df2['athlete_age'] = df2['athlete_age'].astype('string')
df2['athlete_age'] = df2['athlete_age'].str.split('.').str.get(0)
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis = 1)
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2 = df2.rename(columns = {
    'Year of event':'year',
    'Event dates':'race_day',
    'Event name':'race_name',
    'Event distance/length':'race_lenght',
    'Event number of finishers':'race_number_of_finishers',
    'Athlete performance':'athlete_performance',
    'Athlete gender':'athlete_gender',
    'Athlete average speed':'athlete_average_speed',
    'Athlete ID':'athlete_id', 
})
```

```python
df2.head()
```

```python
#Reorde column cause I just dont like how it is being ordered right now
df3 = df2[['race_day','race_name','race_lenght','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed','year']]
```

```python
 # Difference in speed for 50km/50mi male to female
df3.groupby(['race_lenght','athlete_gender'])['athlete_average_speed'].mean()
```

```python
# What age group are the best at 50mi race
df3.query('race_lenght=="50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```

```python
# What age group are the worst at 50mi race
df3.query('race_lenght=="50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True)
```
